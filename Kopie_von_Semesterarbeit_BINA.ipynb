# Korrelation Leerhwohnungsziffer und Bev√∂lkerung


Bev√∂lkerungsdaten bearbeiten

```python
#CSV File Bev√∂lkerung ZH importieren

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Bev√∂lkerung_ZH = pd.read_csv('https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Bev%C3%B6lkerung_ZH.csv')


Bev√∂lkerung_ZH.head()
```

```python
# Pivot funktion nutzen, um Total Jahreszahlen (Kanton) zu erhalten

pivot_table_Bev = Bev√∂lkerung_ZH.pivot_table(index=['INDIKATOR_JAHR'], values=['INDIKATOR_VALUE'], aggfunc='sum')

pivot_table_Bev.head()
```

Leerwohnungdaten *bearbeiten*

```python
#CSV File Bev√∂lkerung ZH importieren

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

Leerhwohnungen_ZH = pd.read_csv('https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Leerhwohnungen_ZH.csv')

Leerhwohnungen_ZH.head()

```

```python
# Pivot funktion nutzen, um Total Jahreszahlen (Kanton) zu erhalten

pivot_table_LW = Leerhwohnungen_ZH.pivot_table(index=['stichjahr'], values=['anz_leerwohnungen'], aggfunc='sum')

pivot_table_LW.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Auswahl der Spalten, die verglichen werden sollen
column1 = 'INDIKATOR_VALUE'
column2 = 'anz_leerwohnungen'  # Name der Spalte in der zweiten Datei

# Berechnen der Korrelation zwischen den ausgew√§hlten Spalten
correlation = pivot_table_Bev[column1].corr(pivot_table_LW[column2])

print(f"Korrelation zwischen {column1} und {column2}: {correlation}")

# Visualisierung der Korrelation (Streudiagramm)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=pivot_table_Bev, x=column1, y=pivot_table_LW[column2])
plt.title(f"Korrelation zwischen {column1} und {column2}")
plt.xlabel(column1)
plt.ylabel(column2)
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


Mietpreise_raw = pd.read_csv('https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Durschn.Mietpreis_NachBaujahr_Zusammengef%C3%BChrt.csv', header=1)

Mietpreise_raw.head()
```

Spalte Kanton an erste Stelle setzen, Vertrauensintervall l√∂schen

```python
cols = Mietpreise_raw.columns.tolist()
cols = [cols[1]] + [col for col in cols if col != cols[1]]
df = Mietpreise_raw[cols]

Mietpreise_merge = df.drop(df.columns[-1], axis=1)

Mietpreise_merge.head()
```

```python
Mietpreise_merge.iloc[:, 2] = Mietpreise_merge.iloc[:, 2].astype(float)
Mietpreise_merge.head()
```
