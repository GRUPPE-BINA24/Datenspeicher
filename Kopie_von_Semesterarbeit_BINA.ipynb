# **Leerstehende Wohnungen (nach Kanton)**

Daten importieren und bereinigen

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Leerstehende_Wohnungen_Kantone.csv', header=None)

dataLeerstehend=df

# zweite Zeile als neue Spaltenüberschriften verwenden
new_columns = dataLeerstehend.iloc[2]
dataLeerstehend.columns = new_columns

# ersten beiden Zeilen entfernen
dataLeerstehend = dataLeerstehend.drop([0, 1, 2])

dataLeerstehend.columns = dataLeerstehend.columns[:2].tolist() + dataLeerstehend.columns[2:].astype(int).tolist()



dataLeerstehend.head()
```

```python
# Spalte Gemeindenummer entfernen
dataLeerstehend = dataLeerstehend.drop(['Gemeindenummer'], axis = 1)
dataLeerstehend.head()
```

```python
#Anzahl der Werte anzeigen
dataLeerstehend.count()
```

Mittelwert, Median und Modus berechne

```python
#Mittelwert berechnen
dataLeerstehend.mean()
```

```python
#Median berechnen
dataLeerstehend.median()
```

```python
#Modus berechnen
dataLeerstehend.mode()
```

Boxplot für die Anzahl Leerstehenden Wohnungen berechnen (nach Jahr)

```python
dataLeerstehend.boxplot(column=[1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023], figsize=(20, 6))


plt.xlabel('Jahr')
plt.ylabel('Anzahl Leerstehende Wohnungen')
plt.title('Boxplot Anzahl Leerstehende Wohnungen (1995 bis 2023)')

```

Boxplot für die Anzahl leerstehender Wohnungen (nach Kanton)

```python
#Transponieren der Tabelle (Kantone zu Spalten, da Boxplot für Spalten ausgelegt) und speicher eines neuen CSVs
dataLeerstehend_Kanton = dataLeerstehend.T

dataLeerstehend_Kanton.to_csv('Kanton_Daten_Spalten.csv', index=False)
```

```python
#neues df erstellen (Import CSV)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


df = pd.read_csv('https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Daten_Kanton_Spalten.csv', header=0)


Data_Kanton=df


Data_Kanton.head()
```

```python
#Daten der Grösse nach sortieren:

# neue Zeile Mittelwert einfügen
Data_Kanton.loc['Mittelwert'] = Data_Kanton.mean()

# Spalten nach dem Mittelwert sortieren und neues df generieren
Kanton_Sorted = Data_Kanton.sort_values(by='Mittelwert', axis=1)

Kanton_Sorted.head()
```

```python
#  "Mittelwert"-Zeile und Jahr-Spalte enfernen, damit sie nicht im Boxplot erscheinen


Kanton_Final = Kanton_Sorted.drop('Jahr', axis=1).drop('Mittelwert')

Kanton_Final.head()
```

```python
#Boxplot erstellen und vertikale Beschriftung einfügen
Kanton_Final.boxplot(figsize=(20, 6))

plt.xticks(rotation='vertical')

plt.ylabel('Anzahl Leerstehende Wohnungen')
plt.title('Boxplot Anzahl Leerstehende Wohnungen (Kantone)')
```
