## Hier werden die Daten von Github eingelesen

```python
# Beispiel-DataFrame, ersetzen Sie dies durch Ihren echten DataFrame
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_excel("https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Mietpreis_BERR.xlsx")
df.head()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Daten laden (Annahme, die Excel-Daten wurden bereits in 'df' geladen)
# Der folgende Code ist eine Fortsetzung und verwendet die Variablen 'df', 'korrigierter_kanton_name' und 'korrigierter_jahr_name' aus deinem vorherigen Code.

# Korrigiere die Namen der Spalten basierend auf der tatsächlichen Formatierung
spaltennamen = df.columns.tolist()
korrigierte_spaltennamen = {str(spalte): spalte for spalte in spaltennamen}  # Konvertiere alle Spaltennamen zu Strings, falls notwendig

# Aktualisiere den Code, um die korrigierten Spaltennamen zu verwenden
korrigierter_kanton_name = korrigierte_spaltennamen.get('Kanton', 'Kanton')
korrigierter_jahr_name = korrigierte_spaltennamen.get('2022', next((name for name in korrigierte_spaltennamen if '2022' in str(name)), None))

# Erstelle den Scatterplot
plt.figure(figsize=(10, 6))  # Setzt die Größe des Diagramms
plt.scatter(df[korrigierter_kanton_name], df[korrigierter_jahr_name], color='blue')  # Erstelle einen Scatterplot mit den korrigierten Spaltennamen
plt.title('Mietpreise nach Kanton im Jahr 2022')  # Setzt den Titel des Diagramms
plt.xlabel(korrigierter_kanton_name)  # Beschriftung der x-Achse
plt.ylabel(f'Mietpreis (CHF pro m²) in {korrigierter_jahr_name}')  # Beschriftung der y-Achse
plt.xticks(rotation=90)  # Drehe die Beschriftungen der x-Achse für eine bessere Lesbarkeit
plt.tight_layout()  # Verbessert das Layout, sodass alles schön aussieht
plt.show()  # Zeigt den Plot

#pfad = "C:\\Users\\raphael.bernhardsgru\\OneDrive - OST\Bilder\Jupyter\Figure1  .jpeg"
#plt.savefig(pfad, format='jpeg')  # Speichert den Plot

pfad = "C:\\Users\\raphael.bernhardsgru\\OneDrive - OST\\Bilder\\Jupyter\\Figure1"
plt.savefig(pfad, format='jpeg')  # Speichert den Plot

plt.close()  # Schließt die Figur
```
