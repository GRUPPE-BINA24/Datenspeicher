## Mietpreisentwicklung nach Kantonen von 2012 bis 2022

```python
import pandas as pd
Dataset = ("https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Mietpreise_nachKantonen_2012_2022_bearbeitet_3ZW.xlsx")
df=pd.read_excel(Dataset)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
kanton = df['Kanton']
mietpreis = df['Mietpreis']
jahr = df['Jahr']

# Liniendiagramm erstellen
plt.figure(figsize=(10,5))
for kanton, jahr, mietpreis in zip(kanton, jahr, mietpreis):
    plt.plot(mietpreis, jahr, marker='o', label=kanton)

plt.xlabel('Mietpreis')
plt.ylabel('Jahr')
plt.legend(ncol=2, loc='center right')
plt.grid(True)
plt.title('Mietpreisentwicklung nach Kantonen von 2012 bis 2022')
plt.show()
```
