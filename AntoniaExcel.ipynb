```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten als DataFrame einlesen
data = {
    "Kanton": ["Zürich"],
    2022: [20.00],
    2021: [19.70],
    2020: [19.20],
    2019: [19.30],
    2018: [19.10],
    2017: [18.70],
    2016: [18.60],
    2015: [18.30],
    2014: [18.80],
    2013: [18.80],
    2012: [18.50]
}

df = pd.DataFrame(data)

# Diagramm erstellen
plt.figure(figsize=(10, 6))
plt.plot(df.columns[1:], df.iloc[0, 1:], marker='o')
plt.title('Temperaturen in Zürich über die Jahre')
plt.xlabel('Jahr')
plt.ylabel('Temperatur (°C)')
plt.grid(True)
plt.show()

```
