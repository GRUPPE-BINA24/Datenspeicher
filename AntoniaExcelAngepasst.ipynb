*Beschreibung der Grafik*:
Die Grafik zeigt die Mietpreisentwicklung für 3-Zimmerwohnungen von bewohnten Wohnungen in Schweizer Kantonen von 2012 bis 2022. Jeder Linie repräsentiert einen Kanton, wobei die Durchschnittsmietpreise pro Quadratmeter (in CHF) über die Jahre aufgetragen sind. Die meisten Kantone zeigen eine allmähliche Zunahme der Mietpreise im betrachteten Zeitraum. Die Legende ist alphabetisch nach den Kantonsnamen sortiert, wobei der Kanton "Schweiz" in schwarzer Farbe hervorgehoben ist. Dies ermöglicht einen direkten Vergleich der Mietpreisentwicklung in den einzelnen Kantonen mit dem schweizerischen Durchschnitt.

*Auffälligkeiten*: Auffällig ist, dass die Mietpreise in den meisten Kantonen tendenziell steigen, wobei einige Kantonen stärkere Steigerungen aufweisen als andere. Einige Gründe für diese Unterschiede könnten wirtschaftliche Faktoren sein, wie die Attraktivität der Region für Unternehmen und Arbeitskräfte, das Angebot und die Nachfrage auf dem Immobilienmarkt, die Infrastruktur- und Verkehrsanbindung sowie die allgemeine wirtschaftliche Entwicklung des Kantons. Regionen mit stärkerem wirtschaftlichem Wachstum, höherer Lebensqualität und besserer Infrastruktur tendieren dazu, höhere Mietpreissteigerungen zu verzeichnen.

*Einschränkungen*: Die Ergebnisse der Studie sind aufgrund methodischer Änderungen ab 2018 in der Erhebungsmethode, Gewichtung und Ausreißererkennung nur bedingt vergleichbar mit den Daten vor 2018. Ebenso sind Vergleiche vor 2015 eingeschränkt, da der Fragebogen angepasst wurde. Die Strukturerhebung berücksichtigt Personen ab 15 Jahren in Privathaushalten der ständigen Wohnbevölkerung, jedoch nicht Personen in Kollektivhaushalten, Diplomaten, internationale Funktionäre und Angehörige. Die Befragten gaben Auskunft über ihre Person, den Haushalt, Haushaltsmitglieder und die Wohnsituation, einschließlich des monatlichen Nettomietzinses von Mieter- und Genossenschaftswohnungen. Teilweise basieren Ergebnisse auf einer Extrapolation mit weniger als 50 Beobachtungen und erfordern eine besonders vorsichtige Interpretation.
