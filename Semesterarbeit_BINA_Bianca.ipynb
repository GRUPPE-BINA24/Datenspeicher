```python
import pandas as pd
import numpy as np

# Daten von der URL laden
url = "https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Mietpreise_LIK_Zins_transponiert.xlsx"
daten = pd.read_excel(url)

# Spaltennamen anpassen, falls erforderlich
spalte1_name = 'Schweiz'
spalte2_name = 'Referenzzinssatz'

# Korrelationskoeffizient berechnen
korrelation = np.corrcoef(daten[spalte1_name], daten[spalte2_name])[0, 1]

print("Korrelationskoeffizient zwischen", spalte1_name, "und", spalte2_name, "ist:", korrelation)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Daten aus Excel-Datei laden
daten = pd.read_excel("https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Mietpreise_LIK_Zins_transponiert.xlsx")

# Spalten auswählen
x = daten['Schweiz']
y = daten['Referenzzinssatz']

# Streudiagramm erstellen
plt.scatter(x, y)
plt.title('Streudiagramm')
plt.xlabel('Mietpreise [CHF]')
plt.ylabel('Referenzzinssatz [Prozent]')
plt.xlim(15, 17.5)
plt.ylim(-1.5, 3)
plt.grid(True)
plt.show()
```

Versucht Dropdown

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

# Daten aus Excel-Datei laden
daten = pd.read_excel("https://raw.githubusercontent.com/GRUPPE-BINA24/Datenspeicher/main/Mietpreise_LIK_Zins_transponiert.xlsx")

# Funktion zum Zeichnen des Streudiagramms mit verbundenen Linien für eine ausgewählte x-Spalte
def plot_selected_x_column(x_column):
    fig, ax = plt.subplots(figsize=(10, 6))  # Größe der Figure anpassen

    # Streudiagramm erstellen
    plt.scatter(daten[x_column], daten['Referenzzinssatz'], label=x_column)
    plt.plot(daten[x_column], daten['Referenzzinssatz'], linestyle='-', linewidth=1)

    plt.title('Streudiagramm mit verbundener Linie')
    plt.xlabel('Mietpreise [CHF]')
    plt.ylabel('Referenzzinsatz [Prozent]')
    plt.ylim(-1.5, 3)
    plt.grid(True)
    plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))  # Legende rechts von der Grafik platzieren
    plt.tight_layout()  # Platz für Achsenbeschriftungen optimieren
    plt.show()

# Dropdown-Menü für die Auswahl der x-Spalte
x_column_dropdown = Dropdown(options=daten.columns, description='x-Spalte wählen:')

# Interaktive Anzeige des Streudiagramms mit Dropdown-Menü
interact(plot_selected_x_column, x_column=x_column_dropdown)
```
